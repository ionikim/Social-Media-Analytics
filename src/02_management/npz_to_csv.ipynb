In this file, we convert the npz files to csv files in order to submit them no neo4j.
--> Problem: they seem to be too big to upload.

```python
import numpy as np
import pandas as pd
import scipy.sparse as sp

# file should be in the same (local) folder as the neo4j folder
```

```python
ictal_1735_1740_adjacency_sparse = np.load('adjacency_sparse/ictal_1735_1740_adjacency_sparse.npz')

# Reconstruct the sparse matrix
matrix = sp.csr_matrix(
    (ictal_1735_1740_adjacency_sparse['data'],
     ictal_1735_1740_adjacency_sparse['indices'],
     ictal_1735_1740_adjacency_sparse['indptr']),
    shape=ictal_1735_1740_adjacency_sparse['shape'])

# Convert to dense and save as CSV in the same folder
df = pd.DataFrame(matrix.toarray())
df.to_csv('adjacency_sparse/ictal_1735_1740_adjacency.csv', index=False)
```

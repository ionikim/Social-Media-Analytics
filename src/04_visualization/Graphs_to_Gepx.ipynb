```python
import pandas as pd
import networkx as nx
from pathlib import Path

def csv_to_full_gexf(nodes_path, edges_path, output_path, directed=False):
    nodes = pd.read_csv("nodes_path") # put your own path
    edges = pd.read_csv("edges_path") # put your own path

    G = nx.DiGraph() if directed else nx.Graph()

    for _, row in nodes.iterrows():
        node_id = str(row["Id"])
        attrs = {
            "label": str(row["Label"]),
            "Label": str(row["Label"]),
            "electrode": str(row["electrode"]),
            "time_idx": int(row["time_idx"]),
            "layer": int(row["layer"]),
            "graph_label": str(row["graph_label"]),
        }
        G.add_node(node_id, **attrs)

    for _, row in edges.iterrows():
        source = str(row["Source"])
        target = str(row["Target"])
        attrs = {
            "source_label": str(row["source_label"]),
            "target_label": str(row["target_label"]),
            "edge_type": str(row["edge_type"]),
            "layer": int(row["layer"]),
            "weight": float(row["weight"]),
        }
        G.add_edge(source, target, **attrs)

    output_path = Path(output_path)
    output_path.parent.mkdir(parents=True, exist_ok=True)
    nx.write_gexf(G, output_path)

    print(f"saved: {output_path}")
    print(f"nodes: {G.number_of_nodes()}")
    print(f"edges: {G.number_of_edges()}")

# ICTAL
csv_to_full_gexf(
    "data/graphs/nodes/graph_ictal_chb01_03_nodes.csv",
    "data/graphs/edges/graph_ictal_chb01_03_edges.csv",
    "data/graphs/gephi/graph_ictal_chb01_03_full.gexf"
)

# INTERICTAL
csv_to_full_gexf(
    "data/graphs/nodes/graph_interictal_chb01_10_nodes.csv",
    "data/graphs/edges/graph_interictal_chb01_10_edges.csv",
    "data/graphs/gephi/graph_interictal_chb01_10_full.gexf"
)
```
